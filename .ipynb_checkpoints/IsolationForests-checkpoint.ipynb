```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import IsolationForest
from ucimlrepo import fetch_ucirepo
from sklearn.preprocessing import StandardScaler
```

```python
# Fetch dataset from UCI repository
air_quality = fetch_ucirepo(id=360)
```

```python
# Convert to DataFrame
data = air_quality.data.features

# Select features
features = data[['CO(GT)', 'C6H6(GT)', 'NOx(GT)', 'NO2(GT)']]

# Drop rows with missing values (-200)
features = features.replace(-200, np.nan)
features = features.dropna()

print(features.shape)
features.head()
```

```python
# Parameters
n_estimators = 100  # Number of trees
contamination = 0.01  # Expected proportion of anomalies
sample_size = 256  # Number of samples used to train each tree
```

```python
# Train Isolation Forest
iso_forest = IsolationForest(n_estimators=n_estimators,
                            contamination=contamination,
                            max_samples=sample_size,
                            random_state=42)
iso_forest.fit(features)
```

```python
# Calculate anomaly scores and classify anomalies
data = data.loc[features.index].copy()
data['anomaly_score'] = iso_forest.decision_function(features)
data['anomaly'] = iso_forest.predict(features)

data['anomaly'].value_counts()
```

```python
# Visualization of the results
plt.figure(figsize=(10, 5))

# Plot normal instances
normal = data[data['anomaly'] == 1]
plt.scatter(normal.index, normal['anomaly_score'], label='Normal')

# Plot anomalies
anomalies = data[data['anomaly'] == -1]
plt.scatter(anomalies.index, anomalies['anomaly_score'], label='Anomaly')
plt.xlabel("Instance")
plt.ylabel("Anomaly Score")
plt.legend()
plt.show()
```

```python
# Visualization of the results
plt.figure(figsize=(5, 5))

# Plot non-anomalies then anomalies
plt.scatter(normal['CO(GT)'], normal['NO2(GT)'], label='Normal')
plt.scatter(anomalies['CO(GT)'], anomalies['NO2(GT)'], label='Anomaly')
plt.xlabel("CO(GT)")
plt.ylabel("NO2(GT)")
plt.legend()
plt.show()
```

### Another example with Bank Transactions

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

# 1. Generate fake transaction data
np.random.seed(42)
n_samples = 300

# Simulate normal transactions
amounts = np.random.normal(loc=50, scale=10, size=n_samples)

# Add a few outliers (unusual high-value transactions)
amounts_with_outliers = np.append(amounts, [200, 250, 300, 400])

df = pd.DataFrame({'amount': amounts_with_outliers})
```

```python
# 2. Fit Isolation Forest
model = IsolationForest(contamination=0.01)  # assume 1% anomalies
model.fit(df[['amount']])

# 3. Predict anomalies
df['anomaly'] = model.predict(df[['amount']])  # -1 = anomaly, 1 = normal
df['score'] = model.decision_function(df[['amount']])

# 4. Visualize
plt.figure(figsize=(10,5))
plt.scatter(df.index, df['amount'], c=df['anomaly'], cmap='coolwarm')
plt.title("Isolation Forest - Anomaly Detection")
plt.xlabel("Transaction Index")
plt.ylabel("Transaction Amount")
plt.grid(True)
plt.show()
```

### Feature Engineering (Per User behavior)

Here we taylor the IsolationForest model to consider the fact that different users may have different spending behaviors. 

We introduce new features that enable us to capture this deferring information inherently. One such feature could be;

- `Amount_Relative`: Amount relative to user account size.
- `Amount z-score`: Amount computed with mean and std of user.
- `Percent balance`: amount relative to account balance.

```python
# Load the CSV
df = pd.read_csv("simulated_transactions.csv")

# Show basic info and preview
print(df.info())
print(df.head())
```

```python
# === Basic Cleaning ===

# Drop duplicates
df = df.drop_duplicates()

# Check for missing values
print(df.isnull().sum())
```

```python
# You can also sort transactions per user
df = df.sort_values(by=['user_id', 'timestamp'])

# Preview again
print(df.head())
```

```python
features = ['amount', 'account_balance', 'amount_to_balance_ratio', 'zscore_amount']
X = df[features].copy()


#Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Initialize and fit the model
iso_model = IsolationForest(n_estimators=100, contamination=0.01, random_state=42)
df['anomaly'] = iso_model.fit_predict(X_scaled)
```

```python
# Count anomalies
print(df['anomaly'].value_counts())

# View a few anomalies
anomalies = df[df['anomaly'] == -1]
print(anomalies.head())
```

```python
df[df['user_id']=='user_113'].head()
```

### Lets visualize per-user anomalies over time using line

```python
# Pick a few sample users to visualize
sample_users = df['user_id'].unique()[11:15]  # adjust number as needed

for user in sample_users:
    user_df = df[df['user_id'] == user].sort_values(by='timestamp')
    plt.figure(figsize=(10, 4))
    plt.plot(user_df['timestamp'], user_df['amount'], label='Amount')
    plt.scatter(user_df[user_df['anomaly'] == -1]['timestamp'],
                user_df[user_df['anomaly'] == -1]['amount'],
                color='red', label='Anomaly', zorder=5)
    plt.title(f"User: {user} - Transaction Amounts with Anomalies")
    plt.xlabel("Timestamp")
    plt.ylabel("Transaction Amount")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Normalize anomaly scores for color mapping
from sklearn.preprocessing import MinMaxScaler

# Add score column if not already added
df['anomaly_score'] = iso_model.decision_function(X_scaled)

# Normalize for colormap
scaler = MinMaxScaler()
df['anomaly_score_scaled'] = scaler.fit_transform(df[['anomaly_score']])

threshold = df['anomaly_score'].quantile(0.01)  # top 2% most anomalous
df['anomaly_flag'] = df['anomaly_score'] < threshold
```

```python
true_anomalies = df[df['anomaly_flag'] == True]
len(true_anomalies)
```

### Now we use our data sofar to train a supervised classifier (XGBoost for more accuracy)

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data Split
# Features and target
X = X_scaled
y = df['anomaly_flag'].astype(int)  # convert boolean to 0/1

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
model.fit(X_train, y_train)
```

```python
#Evaluate Model
y_pred = model.predict(X_test)

print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Normal', 'Anomaly'], yticklabels=['Normal', 'Anomaly'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Feature Importance
xgb.plot_importance(model, importance_type='gain', title='XGBoost Feature Importance (Gain)')
plt.show()
```

```python
# More detailed explanation on feature importance
import shap

explainer = shap.Explainer(model, X_train)
shap_values = explainer(X_test)

shap.summary_plot(shap_values, X_test, feature_names=df.columns[:-3])  # exclude score columns
```
